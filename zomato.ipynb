```python
labels = df['Country'].value_counts().head(5).index
values = df['Country'].value_counts().head(5).values
```

```python
plt.pie(
    values,
    labels=labels,
    autopct='%1.1f%%',
    pctdistance=1.2, 
    labeldistance=1.4  
)

plt.show()
```

```python
rating = df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns = {0: 'Rating Count'})
```

```python
rating
```

```python
plt.figure(figsize=(10,5))
```

```python
sns.barplot(data=rating, x='Aggregate rating', y='Rating Count', hue='Aggregate rating', palette='viridis')
plt.xticks(rotation=90)
plt.show()
```

```python
x = pd.crosstab(df['Country'], df['Has Table booking'], normalize='index')
sns.heatmap(data=x, annot=True, fmt='0.2f', cmap='YlGnBu')
plt.show()
```

```python
x = pd.crosstab(df['Country'], df['Has Online delivery'], normalize='index')
sns.heatmap(data=x, annot=True, fmt='0.2f', cmap='YlGnBu')
plt.show()
```

```python
df[df['Aggregate rating'] == 0]['Country'].value_counts().plot(kind='bar', logy=True)
plt.title("Country that have 0 rating")
plt.xticks(rotation=45)
plt.show()
```

```python
df[df['Aggregate rating'] == 4.9]['Country'].value_counts().plot(kind='bar')
plt.title("Country that have 4.9 rating")
plt.xticks(rotation=90)
plt.show()
```

```python
avg_rating = df[df['Aggregate rating'] > 0].groupby(['Country'])['Aggregate rating'].mean().sort_values(ascending=False)
country = df['Country'].value_counts().index
```

```python
sns.barplot(x=country, y= avg_rating, hue=country, palette='viridis')
plt.title("Avg. rating")
plt.xticks(rotation=90)
plt.show()
```
